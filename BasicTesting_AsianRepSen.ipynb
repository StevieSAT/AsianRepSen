## The Connection between Candidates and Issues they mentioned in Tweets

```python
import pandas as pd
from scipy.stats import chi2_contingency

# create a DataFrame with some sample data
data = pd.read_excel(r'...')  ## make sure the file is not saved as 'Strict XML Open Spreadsheet', the extension is the same. 
df = pd.DataFrame(data)

# create a contingency table of the observed frequencies
contingency_table = pd.crosstab(df['Name'], df['S_Issues'])

# perform a chi-squared test
stat, p, dof, expected = chi2_contingency(contingency_table)
print("Chi-squared statistic:", stat)
print("p-value:", p)
```
